```python
df_final['Date']=df_final['Date_of_Journey'].str.split('/').str[0].astype(int)
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1].astype(int)
df_final['Year']=df_final['Date_of_Journey'].str.split('/').str[2].astype(int)
df_final.head()
```

```python
df_final.drop(['Date_of_Journey'],axis=1,inplace=True)
df_final.head()
```

```python
df_final['Dep_hour']=df_final['Dep_Time'].str.split(':').str[0].astype(int)
df_final['Dep_min']=df_final['Dep_Time'].str.split(':').str[1].astype(int)
df_final.drop(['Dep_Time'],axis=1,inplace=True)
df_final.head()
```

```python
df_final['Arrival_hour']=df_final['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)
df_final['Arrival_min']=df_final['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1].astype(int)
df_final.drop(['Arrival_Time'],axis=1,inplace=True)
df_final.head()
```

```python
df_final['Duration_hour']=df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
df_final['Duration_min']=df_final['Duration'].str.split(' ').str[1].str.split('m').str[0]
df_final.drop(['Duration'],axis=1,inplace=True)
df_final.head()
```
