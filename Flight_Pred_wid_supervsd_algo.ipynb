```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_final=pd.read_csv(r'C:\Users\Ramakrushna das\final_dataset_flighttttt_2.csv')
df_final.head()
```

```python
X=df_final.drop(['Price'],axis=1)
y=df_final['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=10)

```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(X_train,y_train)
linear_model_pred=linear_model.predict(X_test)
from sklearn.metrics import r2_score
linear_accuray=r2_score(linear_model_pred,y_test)
print(linear_accuray)
```

```python
from sklearn.metrics import mean_squared_error
mean_s=mean_squared_error(linear_model_pred,y_test)
print(mean_s)
```

```python
from sklearn.model_selection import cross_val_score
cr_val_reg=LinearRegression()
cross_val_score(cr_val_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
cr_val_reg.fit(X_train,y_train)
cr_val_reg_pred=cr_val_reg.predict(X_test)
cr_val_accuracy=r2_score(cr_val_reg_pred,y_test)
print(cr_val_accuracy)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_regssor=Ridge()
parameters={'alpha':[int(x) for x in np.linspace(1,100,100)]}
grid=GridSearchCV(ridge_regssor,param_grid=parameters,scoring='neg_mean_squared_error',cv=5)
grid.fit(X_train,y_train)

```

```python
grid.best_params_
```

```python
best_estimator=grid.best_estimator_
grid_pred=best_estimator.predict(X_test)
grid_acuu=r2_score(grid_pred,y_test)
print(grid_acuu)
```

```python
from sklearn.linear_model import Lasso
lasso_regssor=Lasso()
grid_lasso=GridSearchCV(lasso_regssor,param_grid=parameters,scoring='neg_mean_squared_error',cv=5)
grid_lasso.fit(X_train,y_train)
```

```python
best_esti_of_lasso=grid_lasso.best_estimator_
```

```python
pred_1=best_esti_of_lasso.predict(X_test)
ac_1=r2_score(pred_1,y_test)
print(ac_1)
```

```python
lasso_regssor.fit(X_train,y_train)
ppp=lasso_regssor.predict(X_test)

acci=r2_score(ppp,y_test)
print(acci)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
dtr_pred=dtr.predict(X_test)
dtr_accuy=r2_score(y_test,dtr_pred)
print(dtr_accuy)
mse=mean_squared_error(y_test,dtr_pred)
print(mse)
```

```python
criterion=['squared_error','friedman_mse','absolute_error']
max_features=['auto','sqrt','log2']
max_depth=[int(x) for x in np.linspace(10,1000,10)]
min_samples_split=[int(x) for x in np.linspace(1,15,15)]
min_samples_leaf=[int(x) for x in np.linspace(1,15,15)]
splitter=['best','random']
random_state=[int(x) for x in np.linspace(1,100,100)]


random_grid={'criterion':criterion,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf,
            'splitter':splitter,
            'random_state':random_state,
            }
print(random_grid)
```

```python
dtrr=DecisionTreeRegressor()
from sklearn.model_selection import RandomizedSearchCV
rdmsearch=RandomizedSearchCV(estimator=dtrr,param_distributions=random_grid,n_iter=100,cv=5,verbose=2,random_state=100,n_jobs=-1)
rdmsearch.fit(X_train,y_train)
```

```python
rdm_best_est=rdmsearch.best_estimator_
```

```python
pred_of_rdm=rdm_best_est.predict(X_test)
acc_of_random=r2_score(pred_of_rdm,y_test)
print(acc_of_random)
```
